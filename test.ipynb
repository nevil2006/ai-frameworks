```python
import torch
```

```python
X = torch.tensor([[1.0], [2.0], [3.0], [4.0]])  # inputs
y = torch.tensor([[2.0], [4.0], [6.0], [8.0]])  # targets (y = 2x)
```

```python
w = torch.randn(1, requires_grad=True)  # weight
b = torch.randn(1, requires_grad=True)  # bias
```

```python
optimizer = torch.optim.SGD([w, b], lr=0.01)

# -------------------------
# TRAINING LOOP
# -------------------------
for epoch in range(100):

    # ---- Forward Pass ----
    y_pred = w * X + b

    # ---- Loss (Mean Squared Error) ----
    loss = ((y_pred - y) ** 2).mean()

    # ---- Backward Pass ----
    loss.backward()

    # ---- Update Weights ----
    optimizer.step()

    # ---- Clear Gradients ----
    optimizer.zero_grad()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")

print("Final weight:", w.item())
print("Final bias:", b.item())
```

This code builds and trains a simple machine learning model using PyTorch. The goal of the model is to learn a linear relationship between input and output values. The dataset provided consists of input values X = [1, 2, 3, 4] and corresponding outputs y = [2, 4, 6, 8], which follow the rule y = 2x. Instead of manually defining this rule, the model learns it automatically.

First, the code imports PyTorch, which is a deep learning library used to create and train models. Then, it defines the dataset using tensors. The input X is a column vector of values, and y is the corresponding output vector. These are the data points the model will learn from.

Next, the model parameters are initialized. These include the weight (w) and bias (b). The weight determines how strongly the input affects the output, and the bias shifts the output value. Both are initialized randomly using torch.randn(), and requires_grad=True is set so that PyTorch can track operations on these variables and compute gradients during training.

An optimizer is then defined using stochastic gradient descent (SGD) with a learning rate of 0.01. The optimizer is responsible for updating the values of w and b in a way that reduces the error of the model.

The core of the program is the training loop, which runs for 100 iterations (epochs). In each epoch, the model performs several steps. First is the forward pass, where predictions are made using the formula y_pred = w * X + b. At the beginning, since w and b are random, the predictions will be incorrect.

Next, the loss is calculated using mean squared error (MSE). This computes the average of the squared differences between predicted values and actual values. A higher loss means the model is performing poorly, while a lower loss indicates better predictions.

After calculating the loss, the backward pass is performed using loss.backward(). This step computes the gradients of the loss with respect to w and b. These gradients indicate how much each parameter should change to reduce the error.

Then, the optimizer updates the parameters using optimizer.step(). It adjusts w and b in the direction that minimizes the loss. The amount of adjustment depends on the gradients and the learning rate.

After updating, optimizer.zero_grad() is called to reset the gradients. This is necessary because PyTorch accumulates gradients by default, and failing to clear them would lead to incorrect updates.

During training, the loss is printed every 10 epochs. You will observe that the loss decreases over time, indicating that the model is learning the correct relationship.

Finally, after training is complete, the code prints the learned values of weight and bias. Ideally, the weight will be close to 2 and the bias close to 0, meaning the model has successfully learned the function y = 2x.

In summary, this code demonstrates the complete machine learning workflow: initializing parameters, making predictions, computing loss, calculating gradients, updating parameters, and iterating until the model learns the correct relationship. This same process is used in training complex deep learning models, including neural networks used in computer vision, natural language processing, and real-time AI systems.
