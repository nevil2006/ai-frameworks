```python
import torch.nn as nn

class MyModel(nn.Module):
    def __init__(self):
        super(MyModel, self).__init__()
        self.linear = nn.Linear(10, 5)
    
    def forward(self, x):
        return self.linear(x)
```

```python
loss_fn=nn.MSELoss()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Create model object
model = MyModel()

# Now optimizer works
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
x = torch.tensor(2.0, requires_grad=True)

y = x**2 + 3*x

y.backward()

print(x.grad)  # 7
```

```python
import torch

# Correct input
x = torch.randn(4, 10)   # 4 samples, 10 features

# Dummy target
y = torch.randn(4, 5)

for epoch in range(10):
    y_pred = model(x)

    loss = loss_fn(y_pred, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    print(loss.item())
```
