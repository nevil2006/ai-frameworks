Problem 3 — Reshape

Create tensor [1,2,3,4,5,6] and reshape to:

(2,3)
(3,2)

```python
import torch

x = torch.tensor([1,2,3,4,5,6])  # correct

x1 = x.reshape(2,3)
print(x1)
```

```python
import torch

x = torch.tensor([1,2,3,4,5,6])  # correct

x1 = x.reshape(3,2)
print(x1)
```

Problem 4 — Autograd

```python
x = torch.tensor(4.0, requires_grad=True)

y=x**3+2*x
y.backward()
print(x.grad)
```
